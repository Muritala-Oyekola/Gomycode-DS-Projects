```python
#start of code
print('Welcome to Python Pizza Deliveries')
size = input('Enter the size you want: \nSmall : S\nMedium : M\nLarge : L\n')
add_pepperoni = input('Do you want Pepperoni: Yes or No')
extra_cheese = input('Do you want extra cheese: Yes or No')
bill = 0

#logical part
#checking if it is small size
if size == 'S':
    bill = bill + 15
    #checking if it is adding pepperoni
    if add_pepperoni == 'Yes':
        bill = bill + 2
    if extra_cheese == 'Yes':
        bill = bill + 1

elif size == 'M':
    bill = bill + 20
    if add_pepperoni == 'Yes':
        bill = bill + 3
    if extra_cheese == 'Yes':
        bill = bill + 1

elif size == 'L':
    bill = bill + 25
    if add_pepperoni == 'Yes':
        bill = bill + 3
    if extra_cheese == 'Yes':
        bill = bill + 1
else:
    print('Choose a valid size')

print('Your total is ',bill)
```
