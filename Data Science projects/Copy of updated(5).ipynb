```python
class Player:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def speak(self):
        print(f"My name is {self.name} and I am {self.age} years old.")

    def run(self):
        # code for running in the game
        print('the player is running')

    def jump(self):
        # code for jumping in the game
        print('the player is jumping')

player1 = Player("John", 25)
player1.speak()
```

```python
#class child(Parent1, Parent2, Parent3)


class SuperPlayer(Player):
    def __init__(self, name, age):
        super().__init__(name, age)

    def fly(self):
        # code for flying in the game
        print('The player is flying')

super_player = SuperPlayer("Jack", 30)
super_player.speak()
super_player.fly()
```
